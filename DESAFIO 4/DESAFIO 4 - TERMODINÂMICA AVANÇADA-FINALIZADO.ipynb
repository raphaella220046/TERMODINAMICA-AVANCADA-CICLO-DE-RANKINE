### A máquina térmica

Uma máquina térmica é qualquer dispositivo que absorve calor ($Q$) e converte parte dessa energia em trabalho ($W$). Infelizmente, apenas parte da energia absorvida como calor pode ser convertida em trabalho por uma máquina térmica. A razão para isso é que o calor, ao fluir, traz consigo entropia, que deve ser de alguma forma eliminada antes que o ciclo possa ser reiniciado. Para se livrar da entropia, toda máquina térmica deve descarregar algum calor residual em seu ambiente. O trabalho produzido pela máquina é a diferença entre o calor absorvido e o calor residual expelido. Sendo assim, a eficiência do sistema é determinada pela relação entre o benefício que é o trabalho produzido e o custo que é o calor absorvido. (SCHROEDER, 2000)

### O ciclo de Rankine

O ciclo de Rankine é um ciclo termodinâmico para aplicação em usinas de energia térmica que visa converter calor em trabalho mecânico. O ciclo recebe esse nome em homenagem ao engenheiro escocês William John Macquorn Rankine, que foi um dos pioneiros no estudo da termodinâmica.

Esse ciclo termodinâmico é composto por quatro etapas principais: aquecimento, expansão, resfriamento e compressão. As etapas ocorrem em quatro componentes principais: a caldeira, a turbina, o condensador e a bomba. A sequência de eventos no ciclo de Rankine é a seguinte (ÇENGEL e BOLES,2014):

1. Aquecimento: O fluido de trabalho, que geralmente é água ou vapor de água, é aquecido em uma caldeira usando uma fonte de calor externa, como a queima de combustível. Nessa etapa, o fluido absorve calor e se transforma em vapor de alta pressão.

2. Expansão: O vapor de alta pressão é enviado para uma turbina, onde expande e realiza trabalho mecânico ao acionar as pás da turbina. A energia térmica do vapor é convertida em energia mecânica.

3. Resfriamento: O vapor expandido e de baixa pressão sai da turbina e entra em um condensador, onde é resfriado e condensado em água líquida. Nessa etapa, o calor é removido do vapor e transferido para um meio de resfriamento, geralmente água fria.

4. Compressão: A água líquida condensada é então pressurizada por meio de uma bomba e enviada de volta para a caldeira para reiniciar o ciclo.

Esse processo de aquecimento, expansão, resfriamento e compressão é cíclico, ou seja, continua a ocorrer repetidamente, permitindo a geração contínua de trabalho mecânico.

### Tipos de ciclo Rankine

Existem diferentes variações do ciclo Rankine, cada uma com suas características e aplicações específicas. A seguir, será apresentado quatro tipos de ciclos Rankine:

1. Ciclo Rankine Simples:
   O ciclo Rankine simples é o mais básico e consiste em quatro componentes principais: uma caldeira, uma turbina, um condensador e uma bomba. O fluido de trabalho, geralmente água, é aquecido na caldeira para produzir vapor de alta pressão, que é expandido na turbina, gerando trabalho mecânico. O vapor então é resfriado e condensado no condensador, retornando à forma líquida e sendo pressurizado novamente pela bomba para reiniciar o ciclo. (ÇENGEL e BOLES,2014 e MORAN et al., 2014)

2. Ciclo Rankine Regenerativo:
   No ciclo Rankine regenerativo, a eficiência térmica é melhorada pela utilização de regeneradores, que são aquecedores adicionais. O vapor de alta pressão é direcionado para o regenerador, onde é pré-aquecido pelo vapor de baixa pressão que está saindo da turbina. Isso reduz a quantidade de calor necessário para aquecer o vapor na caldeira, aumentando a eficiência geral do ciclo. (ÇENGEL e BOLES, 2014)

3. Ciclo Rankine com Reaquecimento:
   O ciclo Rankine com reaquecimento é semelhante ao ciclo Rankine simples, mas com a adição de uma etapa intermediária que reaquece o vapor. Após a expansão parcial na turbina, o vapor é direcionado para um reaquecedor, onde é aquecido novamente antes de entrar na turbina para a expansão final. Isso reduz as perdas de irreversibilidade e melhora a eficiência do ciclo. (ÇENGEL e BOLES, 2014)

4. Ciclo Rankine Supercrítico:
   O ciclo Rankine supercrítico é projetado para operar com fluidos acima de seu ponto crítico. Nesse estado, o fluido não passa por uma transição líquido-vapor definida, o que permite que ele alcance eficiências mais altas do que os ciclos subcríticos. O uso de fluidos supercríticos, como o CO2, tem sido explorado para aplicações de alta eficiência. (LIU et al., 2018)
   
Existem outras configurações como o ciclo de Rankine combinado, o ciclo de Rankine binário e o ciclo de Rankine Orgânico (ORC) que implementam melhorias no sistema, podendo ser aplicadas dependendo das condições específicas de operação e dos objetivos de eficiência.

O ciclo de Rankine ainda pode ser classificado em dois tipos principais: ciclo de Rankine aberto e ciclo de Rankine fechado.

1. Ciclo de Rankine Aberto:

No ciclo de Rankine aberto, o fluido de trabalho é aquecido na caldeira e então expandido em uma turbina, produzindo trabalho. Após a expansão na turbina, o fluido de trabalho é liberado para o ambiente, com o lançamento de vapor na atmosfera.

O ciclo de Rankine aberto é comumente usado em usinas termelétricas convencionais, onde a água é o fluido de trabalho e a fonte de calor é fornecida por combustíveis fósseis, como carvão, óleo ou gás natural. A principal desvantagem desse ciclo é o desperdício de água, uma vez que a maior parte do fluido de trabalho não é reaproveitada. (MORAN et al., 2014)

2. Ciclo de Rankine Fechado:

No ciclo de Rankine fechado, o fluido de trabalho é aquecido na caldeira, passa por uma turbina para produzir trabalho e é então resfriado e condensado em um condensador. Após a condensação, o fluido de trabalho é recomprimido e retorna à caldeira para ser aquecido novamente, reiniciando o ciclo.

O ciclo de Rankine fechado é amplamente utilizado em usinas termelétricas, onde o fluido de trabalho pode ser água, mas também podem ser utilizados fluidos orgânicos de baixo ponto de ebulição, conhecidos como Fluidos de Trabalho Orgânicos (FWO). Esses FWOs possuem vantagens, como menor consumo de água e maior eficiência térmica em comparação com o ciclo de Rankine aberto. (ÇENGEL e BOLES, 2014)

O calor é uma forma de energia que pode ser convertida em trabalho. A queima do combustível no ciclo de Rankine libera calor para que o sistema termodinâmico entre em operação.

A variação de calor é dada pelo produto entre a massa $M$, o calor específico ($C$) e a variação de temperatura ($T$).

$$\Delta Q = M \cdot C \cdot \Delta T$$

O calor pode ser calculado numericamente por meio da integral

$$ Q = M \int_{T_f}^{T_0} C_p(T) dT $$

O calor também pode ser obtido pela diferença de entalpia entre o estado final e o estado inicial do sistema, ou seja $$Q = \Delta H$$

Para calcular a quantidade de massa que uma determinada quantidade de calor consegue esquentar, pode-se isolar o $m$ da equação e fazer a substituição em $Q$

$$ M = \Delta H\int_{T_f}^{T_0} C_p(T) dT $$

### Modelagem do ciclo de Rankine

```python
# Verificação das unidades de medida
# pm.config

# Configurações de pressão

pm.config["unit_pressure"] = "kPa" # unidade de medida
pm.config["def_p"] = 100 # pressão padrão

# Requerimento das propriedades da água por meio da biblioteca PyroMat

mp_water = pm.get("mp.H2O")
```

#### BOMBA

Assume-se que a água do condensador recebida pela bomba é um líquido saturado, então x=0.

```python
p1 = 10 # kPa é a pressão do condensador
p2 = 2500 # kPa é a pressão da caldeira
v = 1/mp_water.ds(p=p1)[0] # o volume específico é o inverso da densidade para o líquido saturado
```

Trabalho exigido pela bomba é o volume específico multiplicado pela variação de pressão $W = v \cdot \Delta p$

```python
# Assumindo que o líquido é incompressível e não varia com o aumento da pressão

w_p = v*(p2-p1)

print(f"O trabalho requerido pela bomba é {round(float(w_p),1)} kJ/kg")
```

A entalpia é calculada com base no líquido saturado, então x=0

```python
h1 = mp_water.hs(p=p1)[0] # entalpia para o líquido saturado

h2 = h1+w_p

print(f"A entalpia h2 = {round(float(h2),1)} kJ/kg")
```

#### CALDEIRA

A água que sai da caldeira é um vapor saturado, então x=1

```python
p3 = p2 # pressão constante
h3 = mp_water.hs(p=p3)[1] # entalpia para o vapor saturado
s3 = mp_water.ss(p=p3)[1] # entropia para o vapor saturado

# O calor da caldeira é igual a diferença de entalpia

q_H = h3-h2

print(f"A entrada de calor pela caldeira é {round(float(q_H),1)} kJ/kg")
```

#### TURBINA

A água que sai da caldeira é um vapor saturado, então x=1

```python
p4 = p1 # pressão constante
s4 = s3 # entropia é constante (isentrópico)
x = mp_water.T_s(s=s4, p=p4, quality=True)[1] # qualidade da água multifásica (líquido-vapor)
h4 = mp_water.h(p=p4, x=x) # entalpia do fluido após passar pela turbina

# O trabalho é a diferença de entalpia
w_t = h3-h4

print(f"O Trabalho gerado pela turbina é {round(float(w_t),1)} kJ/kg")
```

#### CONDENSADOR

Calcular o calor liberado pelo condensador ao resfriar a água

```python
# O calor liberado é a diferença de entalpia

q_L = h4 - h1

print(f"O calor liberado pelo condensador é {round(float(q_L),1)} kJ/kg")
```

#### EFICIÊNCIA TÉRMICA

Equação de eficiência térmica

$$\large \eta_{th} = \frac{W_{net}}{q_H}$$

A eficiência é dada pelo trabalho que o fluido exerce na turbina subtraído do trabalho necessário para girar a bomba que pressuriza o fluido ($W_{net}$) dividido pelo calor transferido ao fluido de trabalho pela caldeira ($q_H$).

```python
eta_th = (w_t - w_p) / q_H*100

print(f"A eficiência térmica é {round(float(eta_th),1)} %")
```

### A escolha do ciclo de Rankine

Para a escolha do ciclo de Rankine, foram testados três tipos de ciclo e comparada a eficiência de cada um. Os ciclos testados são: Ciclo de Rankine simples, Ciclo de Rankine Regenerativo e Ciclo de Rankine com Reaquecimento

#### Ciclo de Rankine Simples

Bomba:
A função da bomba é aumentar a pressão do fluido de trabalho antes de entrar na caldeira. Isso é feito aplicando trabalho mecânico à bomba, que consome energia. Sendo assim, na bomba, o trabalho é fornecido ao sistema e nenhum calor é transferido. O fluido de trabalho é comprimido, aumentando sua pressão e preparando-o para entrar na caldeira.

Caldeira:
Na caldeira, o fluido de trabalho de alta pressão e alta temperatura entra e é aquecido pela adição de calor. O calor é fornecido por uma fonte externa, neste caso a queima do tolueno e ciclohexano. O fluido de trabalho absorve calor e sofre uma mudança de fase, de líquido para vapor. Nesse processo, ocorre um aumento significativo de entalpia e a energia térmica é convertida em energia interna. O trabalho realizado na caldeira é zero, pois a pressão é constante.

Turbina:
Após sair da caldeira, o vapor de alta pressão é direcionado para a turbina. Na turbina, o vapor expande, passando através de palhetas em um rotor. À medida que o vapor se expande, seu calor específico diminui e sua pressão e temperatura caem. A queda de pressão e temperatura do vapor faz com que ele realize trabalho mecânico na turbina. Esse trabalho é convertido em energia mecânica, que pode ser usado para acionar um gerador elétrico ou outra máquina. A energia térmica do vapor é convertida em trabalho na turbina.

Condensador:
Após sair da turbina, o vapor é direcionado para o condensador. No condensador, o vapor é resfriado e condensado em líquido novamente. Isso é feito através da transferência de calor para um meio de resfriamento, como água fria. Durante o processo de condensação, o fluido de trabalho perde calor e diminui sua entalpia. O trabalho realizado no condensador é zero, pois a pressão é constante.

```python
# Definição da pressão no condensador e na caldeira

p1_rc = 3.15
p2_rc = 5000
```

```python
# BOMBA
# Definição do volume específico e trabalho

v_rc = 1/mp_water.ds(p=p1_rc)[0]
w_p_rc = v_rc*(p2_rc-p1_rc)
print(f"O trabalho requerido pela bomba é {round(float(w_p_rc),1)} kJ/kg")
```

```python
# Definição da temperatura, entalpia e entropia nesse ponto

T1_rc = mp_water.Ts(p=p1_rc)[0]
h1_rc = mp_water.hs(p=p1_rc)[0]
s1_rc = mp_water.ss(p=p1_rc)[0]
s2_rc= s1_rc
h2_rc = h1_rc+w_p_rc
T2_rc = mp_water.T_h(p=p2_rc,h=h2_rc)

print(f"A entalpia h2 é = {round(float(h2_rc),1)} kJ/kg")
```

```python
# CALDEIRA
# A água deixa a caldeira na forma de vapor saturado, assume-se que a pressão é isobárica nesse ponto

p3_rc = p2_rc
h3_rc = mp_water.hs(p=p3_rc)[1]

# O calor é dado pela diferença de entalpia

q_H_rc = h3_rc-h2_rc

print(f"A entrada de calor na caldeira é {round(float(q_H_rc),1)} kJ/kg")
```

```python
# Definição da temperatura e entropia nesse ponto

T3_rc = mp_water.Ts(p=p3_rc)
s3dash_rc = mp_water.ss(p=p3_rc)[0]
T3dash_rc = T3_rc
s3_rc = mp_water.ss(p=p3_rc)[1]
```

```python
# TURBINA
# Quando o vapor saturado passa pela turbina, a pressão decai para pressão do condensador

p4_rc = p1_rc

# A entalpia é constante (isentrópica)
s4_rc = s3_rc

# Definição da temperatura e entropia nesse ponto

T4_rc, x_rc = mp_water.T_s(s=s4_rc,p=p4_rc, quality=True) # em que x_rc determina a qualidade da água multifásica
h4_rc = mp_water.h(p=p4_rc,x=x_rc)

# O trabalho da turbina
w_t_rc = h3_rc-h4_rc
print(f"O Trabalho gerado pela turbina é {round(float(w_t_rc),1)} kJ/kg")
```

```python
# CONDENSADOR

q_L_rc = h4_rc-h1_rc
print(f"O calor liberado pelo condensador é {round(float(q_L_rc),1)} kJ/kg")
```

```python
# A eficiência do sistema

eta_th_rc = (w_t_rc-w_p_rc)/q_H_rc*100
print(f"A eficiência térmica do ciclo de Rankine simples é {round(float(eta_th_rc),1)}%")
```

#### Ciclo de Rankine Ideal com Reaquecimento

Esse ciclo apresenta a turbina em duas etapas, uma de baixa pressão e outra de alta pressão, foi considerada a pressão na turbina de baixa pressão com a taxa de 20% em relação à turbina de alta pressão

```python
# Definição da pressão no condensador e na bomba

p1_ra = 3.15
p2_ra = 5000

# BOMBA DE BAIXA PRESSÃO

v_ra = 1/mp_water.ds(p=p1_ra)[0]

w_p_ra = v_ra*(p2_ra-p1_ra)

print(f"O volume específico é {round(float(v_ra),5)} m^3/kg")
print(f"O trabalho requerido na bomba 1 é {round(float(w_p_ra),1)} kJ/kg")
```

```python
# A entalpia nos pontos 1 e 2

h1_ra = mp_water.hs(p=p1_ra)[0]
h2_ra = h1_ra+w_p_ra
print(f"A entalpia h2 é {round(float(h2_ra),1)} kJ/kg")
```

```python
# pressão, temperatura, entalpia e entropia nos pontos 3 e 4

p3_ra = p2_ra
T3_ra = 550
h3_ra = mp_water.h(p=p3_ra, T=T3_ra)
s3_ra = mp_water.s(p=p3_ra, T=T3_ra)

p4_ra = 1000
s4_ra = s3_ra
T4_ra, x4_ra = mp_water.T_s(s=s4_ra, p=p4_ra, quality=True)
h4_ra = mp_water.h(x=x4_ra, p=p4_ra)
```

```python
# pressão, temperatura, entalpia e entropia nos pontos 5 e 6

p5_ra = p4_ra
T5_ra = T3_ra
h5_ra = mp_water.h(p=p5_ra, T=T5_ra)
s5_ra = mp_water.s(p=p5_ra, T=T5_ra)

p6_ra = p1_ra
s6_ra = s5_ra
T6_ra, x6_ra = mp_water.T_s(s=s6_ra, p=p6_ra, quality=True)
h6_ra = mp_water.h(x=x6_ra, p=p6_ra)

w_LPt_ra = h5_ra-h6_ra

print(f"O trabalho gerado pela turbina de alta pressão é {round(float(w_LPt_ra),1)} kJ/kg")
print(f"A saída total de trabalho pela turbina é {round(float(w_HPt_ra+w_LPt_ra),1)} kJ/kg")
```

```python
# BOILER

q_H_ra = (h3_ra-h2_ra)+(h5_ra-h4_ra)
print(f"A entrada de calor pela caldeira é {round(float(q_H_ra),1)} kJ/kg")
```

A eficiência no ciclo de Rankine com reaquecimento é calculada por

$$\frac{W}{Q} = \frac{(h_1-h_2)+(h_3-h_4)-w_p}{(h_1-h_5)+(h_3-h_2)-w_p}$$

```python
# Representação gráfica

s1_ra = mp_water.ss(p=p1_ra)[0]
T1_ra = mp_water.Ts(p=p1_ra)[0]
s2_ra = s1_ra

p_ra = np.linspace(1,20000,1000)
T_ra = mp_water.Ts(p=p_ra)
s_ra = mp_water.ss(p=p_ra)

h2dash_ra = mp_water.hs(p=p2_ra)[0]
s2dash_ra = mp_water.ss(p=p2_ra)[0]
T2dash_ra = mp_water.Ts(p=p2_ra)[0]

h3dash_ra = mp_water.hs(p=p2_ra)[1]
s3dash_ra = mp_water.ss(p=p2_ra)[1]
T3dash_ra = T2dash_ra

T2_ra = mp_water.T_h(h=h2_ra,p=p2_ra)

plt.figure(figsize=(15,10))
plt.title('Diagrama T-S Ciclo de Rankine com Reaquecimento', size = 30)
plt.plot(s_ra[0],T_ra, 'b')
plt.plot(s_ra[1],T_ra,'r')
plt.ylabel('Temperatura (K)', size = 25)
plt.xlabel('Entropia (kJ/kgK)', size = 25)
#plt.xlim(-3,10)
#plt.ylim(200,800)
plt.plot([s1_ra, s2_ra, s2dash_ra, s3dash_ra, s3_ra, s4_ra, s5_ra, s6_ra, s1_ra],[T1_ra, T2_ra, T2dash_ra, T3dash_ra, T3_ra, T4_ra, T5_ra, T6_ra, T1_ra], 'black', linewidth=3)

plt.text(s1_ra-.1,T1_ra,f'(1)\nT = {round(float(T1_ra),2)} K\nh = {round(float(h1_ra),1)} kJ/kg\n s = {round(float(s1_ra),3)} kJ/kgK',
    ha='right',backgroundcolor='white')
plt.text(1.6,330,f'(2)\nT = {round(float(T2_ra),2)} K\nh = {round(float(h2_ra),1)} kJ/kg',
    ha='left',backgroundcolor='white')
plt.text(s2dash_ra-.15,T2dash_ra,f"(2')\nT = {round(float(T2dash_ra),2)} K\nh = {round(float(h2dash_ra),1)} kJ/kg \ns = {round(float(s2dash_ra),3)} kJ/kgK",
    ha='right',backgroundcolor='white')
plt.text(s3dash_ra-.1,T3dash_ra-80,f"(3')\nh = {round(float(h3dash_ra),1)} kJ/kg \ns = {round(float(s3dash_ra),3)} kJ/kgK",
    ha='right',backgroundcolor='white')
plt.text(6.3,T3_ra-50,f'(3)\nT = {round(float(T3_ra),2)} K\nh = {round(float(h3_ra),1)} kJ/kg',
    ha='right',backgroundcolor='white')
plt.text(s4_ra-.1,T4_ra-80,f'(4)\nT = {round(float(T4_ra),2)} K\nh = {round(float(h4_ra),1)} kJ/kg \ns = {round(float(s4_ra),3)} kJ/kgK',
    ha='right',backgroundcolor='white')
plt.text(s5_ra+.1,T5_ra-70,f'(5)\nT = {round(float(T4_ra),2)} K\nh = {round(float(h4_ra),1)} kJ/kg \ns = {round(float(s4_ra),3)} kJ/kgK',
    ha='left',backgroundcolor='white')
plt.text(s6_ra+.1,T6_ra,f'(6)\nT = {round(float(T4_ra),2)} K\nh = {round(float(h4_ra),1)} kJ/kg',
    ha='left',backgroundcolor='white')

#plt.savefig("Diagrama T-S ciclo de Rankine com reaquecimento")
plt.show()
```

#### INTERPRETAÇÃO

Bomba:
Assim como no ciclo básico de Rankine, a bomba comprime a água, aumentando sua pressão. Essa etapa é representada por uma linha quase vertical no diagrama T-S, onde a entropia geralmente não varia significativamente. A temperatura da água pode aumentar ligeiramente devido ao trabalho realizado pela bomba.

Caldeira:
Na caldeira, a água é aquecida a uma temperatura elevada, transformando-se em vapor. Nessa etapa, o vapor se expande na primeira etapa da turbina, até um ponto chamado ponto de reaquecimento. Nesse ponto, o vapor é retirado da primeira expansão e é enviado para um reaquecedor, onde é aquecido novamente antes de entrar na segunda etapa da turbina. O vapor passa por um processo de expansão adiabática durante a primeira etapa da turbina e uma expansão isobárica durante a segunda etapa. Essas etapas são representadas por linhas inclinadas no diagrama T-S, onde a temperatura e a entropia mudam ao longo das expansões.

Reaquecimento:
No reaquecedor, o vapor é aquecido novamente antes de entrar na segunda etapa da turbina. Essa etapa é representada por uma linha quase horizontal no diagrama T-S, pois a temperatura do vapor aumenta enquanto a entropia permanece praticamente constante.

Turbina:
Na segunda etapa da turbina, o vapor expandido realiza trabalho mecânico, resultando em uma queda de temperatura e entropia. Essa etapa é representada por uma linha inclinada no diagrama T-S.

Condensador:
Após a segunda etapa da turbina, o vapor é direcionado para o condensador, onde é resfriado e condensado. Essa etapa é representada por uma linha quase horizontal no diagrama T-S, indicando a queda de temperatura enquanto a entropia permanece praticamente constante.

#### Ciclo de Rankine Ideal Regenerativo

Esse ciclo apresenta duas bombas, uma de baixa pressão e outra de alta pressão, foi considerada a pressão na bomba de baixa pressão com a taxa de 10% em relação à bomba de alta pressão

```python
# Definição da pressão no condensador e na bomba 1

p1_re = 5000
p2_re = 3.15

# BOMBA DE BAIXA PRESSÃO

v1_re = 1/mp_water.ds(p=p1_re)[0]

w_pump1_re = v1_re*(p1_re-p2_re)
print(f"O trabalho requerido na bomba 1 é {round(float(w_pump1_re),1)} kJ/kg")
```

```python
# pressão, temperatura, entalpia e entropia nos pontos 5 e 6

p5_re = 500
T1_re = 550
s1_re = mp_water.s(p=p1_re,T=T1_re)
s5_re = s1_re
T5_re, x5_re = mp_water.T_s(p=p5_re,s=s5_re,quality='True')
h5_re = mp_water.h(p=p5_re, x=x5_re)

p6_re = p5_re
s6_re = mp_water.ss(p=p6_re)[0]
T6_re = mp_water.Ts(p=p6_re)
h6_re = mp_water.hs(p=p6_re)[0]
v6_re = 1/mp_water.ds(p=p6_re)[0]
```

```python
# AQUECEDOR DA ÁGUA DE ALIMENTAÇÃO

p3_re = p2_re
h3_re = mp_water.hs(p=p3_re)[0]
```

A equação de conservação de energia para o FWH (*Feed Water Heat*) é $$y(h_6)+(1-y)h_2=h_3$$
Para encontrar a fração $y$ referente entalpia $$y = \frac{h_2-h_3}{h_2-h_6}$$

```python
# Entalpia nos pontos 1 e 2

h1_re = mp_water.h(p=p1_re, T=T1_re)[0]
s2_re = s1_re
T2_re, x2_re = mp_water.T_s(p=p2_re,s=s2_re,quality=True)
h2_re = mp_water.h(p=p2_re,x=x2_re)

y_re = (h2_re-h3_re)/(h2_re-h6_re)

p7_re = p1_re
s7_re = s6_re
w_bfp = v6_re*(p7_re-p6_re)
h7_re = h6_re+w_bfp
T7_re = mp_water.T_s(s=s7_re,p=p7_re)

# TURBINA

# O trabalho realizado na turbina é a diferença de entalpia entre os pontos 5, 6 e 7
w_t_re = h5_re - y_re*h6_re - (1-y_re)*h7_re
print(f"Trabalho gerado pela turbina é {round(float(w_t_re),1)} kJ/kg")
```

```python
# BOMBA DE ALTA PRESSÃO

p4_re = p5_re
v3_re = 1/mp_water.ds(p=p3_re)[0]

w_pump2_re = v3_re*(p4_re-p3_re)
print(f"O trabalho requerido pela bomba 2 é {round(float(w_pump2_re),1)} kJ/kg")
print(f"O trabalho total geral pelas bombas é {round(float(w_pump1_re+w_pump2_re),1)} kJ/kg")
```

```python
# CALDEIRA

p8_re = p1_re
p9_re = p8_re
h4_re = h3_re + w_pump2_re
q_H_re = h5_re-h4_re
print(f"A entrada de calor na caldeira é {round(float(q_H_re),1)} kJ/kg")

h8_re = mp_water.hs(p=p8_re)[0]
q_L_re = h1_re - h8_re
print(f"O calor liberado pelo condensador é {round(float(q_L_re),1)} kJ/kg")
```

A eficiência no ciclo de Rankine com reaquecimento pode ser obtida por

$$e = 1-\frac{Q_c}{Q_h}$$

Em que $Q_c$ corresponde ao calor absorvido e $Q_h$ representa o calor expelido pelo sistema

```python
# A eficiência do sistema

# Cálculo de massa extraída
m_re = (h6_re - h4_re) / (h5_re - h4_re)

# Trabalho realizado pelo ciclo é a diferença entre o trabalho realizado na turbina e o trabalho para acionar as bombas
w_net = w_t_re - w_pump2_re - w_bfp

# Cálculo do calor absorvido e expelido do sistema
qh_re = h1_re - h7_re
qc_re = (1 - m_re) * (h2_re - h3_re)

# Determinação da eficiência
eta_th_re = (1 - (qc_re / qh_re)) * 100

print('A eficiência térmica do ciclo de Rankine regenerativo é ',round(eta_th_re[0],1),'%')
```

```python
# Definição de temperatura e entropia nos pontos 3, 4, 8 e 9

T3_re = mp_water.Ts(p=p3_re)
s3_re = mp_water.ss(p=p3_re)[0]

s4_re = s3_re
T4_re = mp_water.T_s(s=s4_re,p=p4_re)

s8_re = mp_water.ss(p=p8_re)[0]
T8_re = mp_water.Ts(p=p8_re)

s9_re = mp_water.ss(p=p8_re)[1]
T9_re = T8_re

p_re = np.linspace(1,20000,1000)
T_re = mp_water.Ts(p=p_re)
s_re = mp_water.ss(p=p_re)

plt.figure(figsize=(15,10))
plt.font = {'family':'Arial', 'size': 14}
plt.rc('font', **plt.font)
plt.title('Diagrama T-S Ciclo de Rankine Regenerativo', size = 30)
plt.plot(s_re[0],T_re, 'b')
plt.plot(s_re[1],T_re,'r')
plt.ylabel('Temperature (K)', size = 25)
plt.xlabel('Entropia (kJ/kgK)', size = 25)
#plt.xlim(0,9)
#plt.ylim(200,800)

s_re = [s1_re, s5_re, s2_re, s3_re, s4_re, s6_re, s7_re, s8_re, s9_re, s1_re]
T_re = [T1_re, T5_re, T2_re, T3_re, T4_re, T6_re, T7_re, T8_re, T9_re, T1_re]
plt.plot(s_re, T_re, 'k',linewidth=3)
plt.plot([s5_re ,s6_re],[T5_re, T6_re],'g',linewidth=3)

plt.text(s1_re+0.5,T1_re+15,f'(1)\nT = {round(float(T1_re),2)} K\nh = {round(float(h1_re),1)} kJ/kg\n s = {round(float(s1_re),3)} kJ/kgK',
    ha='right',backgroundcolor='white')
plt.text(s5_re+0.1,T5_re, f'(5)\nT = {round(float(T5_re),2)} K\nh = {round(float(h5_re),1)} kJ/kg\n s = {round(float(s1_re),3)} kJ/kgK',
    ha='left',backgroundcolor='white')
plt.text(s2_re+0.1,T2_re, f'(2)\nT = {round(float(T2_re),2)} K\nh = {round(float(h2_re),1)} kJ/kg\n s = {round(float(s2_re),3)} kJ/kgK',
    ha='left',backgroundcolor='white')
plt.text(s3_re-0.3,T3_re, f'(3)\nT = {round(float(T3_re),2)} K\nh = {round(float(h3_re),1)} kJ/kg\n s = {round(float(s3_re),3)} kJ/kgK',
    ha='left',backgroundcolor='white')
plt.text(s4_re-0.3,T4_re+55, f'(4)\nT = {round(float(T4_re),2)} K\nh = {round(float(h4_re),1)} kJ/kg\n s = {round(float(s4_re),3)} kJ/kgK',
    ha='left',backgroundcolor='white')
plt.text(s6_re-0.8,T6_re, f'(6)\nT = {round(float(T6_re),2)} K\nh = {round(float(h6_re),1)} kJ/kg\n s = {round(float(s6_re),3)} kJ/kgK',
    ha='left',backgroundcolor='white')
plt.text(s7_re-0.8,T7_re+50, f'(7)\nT = {round(float(T7_re),2)} K\nh = {round(float(h7_re),1)} kJ/kg\n s = {round(float(s7_re),3)} kJ/kgK',
    ha='left',backgroundcolor='white')
plt.text(s8_re-0.8,T8_re+10, f'(8)\nT = {round(float(T8_re),2)} K\nh = {round(float(h8_re),1)} kJ/kg\n s = {round(float(s8_re),3)} kJ/kgK',
    ha='left',backgroundcolor='white')
plt.text(s9_re+0.2,T9_re-40, f'(9)\nT = {round(float(T9_re),2)} K\n s = {round(float(s9_re),3)} kJ/kgK',
    ha='left',backgroundcolor='white')
plt.text((s5_re+s6_re)/2,(T5_re)+10,'m_re')
plt.text((s3_re+s2_re)/2+0.3,(T3_re)+10,'1-m_re')

#plt.savefig("Diagrama T-S ciclo de Rankine regenerativo")
plt.show()
```

#### INTERPRETAÇÃO

Bomba:
Assim como no ciclo básico de Rankine, a bomba comprime a água, aumentando sua pressão. Essa etapa é representada por uma linha quase vertical no diagrama T-S, onde a entropia geralmente não varia significativamente. A temperatura da água pode aumentar ligeiramente devido ao trabalho realizado pela bomba.

Caldeira:
Na caldeira, a água é aquecida a uma temperatura elevada, transformando-se em vapor. Nessa etapa, o vapor é enviado para a turbina, onde realiza trabalho mecânico. A expansão do vapor na turbina é representada por uma linha inclinada no diagrama T-S, onde ocorre uma queda de temperatura e entropia.

Regenerador:
Após a expansão na turbina, parte do vapor de exaustão é desviado para o regenerador. No regenerador, o vapor de exaustão aquece a água de alimentação antes que ela entre na caldeira. Essa etapa é representada por uma linha quase horizontal no diagrama T-S, onde a entropia permanece praticamente constante enquanto a temperatura da água de alimentação aumenta.

Turbina:
Após o pré-aquecimento no regenerador, a água de alimentação aquecida entra na caldeira e é convertida em vapor de alta pressão novamente. Esse vapor passa pela turbina, onde realiza trabalho mecânico. A expansão do vapor na turbina é representada por uma segunda linha inclinada no diagrama T-S, onde ocorre uma nova queda de temperatura e entropia.

Condensador:
Após a segunda expansão na turbina, o vapor é direcionado para o condensador, onde é resfriado e condensado. Essa etapa é representada por uma linha quase horizontal no diagrama T-S, indicando a queda de temperatura enquanto a entropia permanece praticamente constante.
